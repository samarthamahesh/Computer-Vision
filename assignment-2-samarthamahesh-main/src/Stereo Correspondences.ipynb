```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Helper Functions

```python
def import_imgs(idx):
    '''
    Function to import a set of images
    
    Input
    -----
    idx: Index of the image
    
    Output
    ------
    imgs: Array of images
    
    '''
    imgs = []
    idx = str(idx)
    ilen = len(idx)
    path = '../images/Stereo Correspondence/'
    
    for i in os.listdir(path):
        if i[:ilen] == idx:
            img = cv2.imread(path+i)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            imgs.append(img)
            
    return imgs
```

```python
def get_keypoints(im):
    '''
    Function to find keypoints and descriptors using feature detector and descriptor (SIFT)
    
    Input
    -----
    im: Image
    
    Output
    ------
    kp: Keypoints
    des: Descriptor
    
    '''
    sift = cv2.SIFT_create()
    kp, des = sift.detectAndCompute(im, None)
    
    return kp, des

def matching(kp1, des1, kp2, des2, ratio=0.5):
    '''
    Function tp find matches between two images using SIFT outputs
    
    Input
    -----
    kp1: Keypoints of image 1
    des1: Descriptor of image 1
    kp2: Keypoints of image 2
    des2: Descriptor of image 2
    ratio: Paramater to decide the match based on distances
    
    Output
    ------
    new_matches: Matches obtained between two images
    
    '''
    FLANN_INDEX_KDTREE = 1
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks=50) 

    flann = cv2.FlannBasedMatcher(index_params, search_params)

    matches = flann.knnMatch(des1, des2, k=2)

    new_matches = []
    for i, (m, n) in enumerate(matches):
        if m.distance < ratio * n.distance:
            new_matches.append([m.queryIdx, m.trainIdx])
    
    return np.asarray(new_matches)

def get_match_coord(matches, kp1, kp2):
    '''
    Function to get coordinates of the matches obtained
    
    Input
    -----
    matches: Matches indices
    
    Output
    ------
    coord1: Coordinates of the match points in image 1
    coord2: Coordinates of the match points in image 2
    
    '''
    coord1 = []
    coord2 = []
    for i, j in matches:
        coord1.append(kp1[i].pt)
        coord2.append(kp2[j].pt)
        
    return np.asarray(coord1), np.asarray(coord2)
```

## 2.1 You are given two images of the same scene, taken from different view points. Perform Intensity window-based correlation on the given pair of images.

```python
def correlation(patch1, patch2):
    '''
    Function to find correlation between two patches
    
    Input
    -----
    patch1: Patch 1
    patch2: Patch 2
    
    Output
    ------
    corr: Correlation between patch1 and patch2
    
    '''
    patch1 = np.float32(patch1)
    patch2 = np.float32(patch2)
    
    num = np.mean((patch1 - np.mean(patch1))*(patch2 - np.mean(patch2)))
    den = np.std(patch1)*np.std(patch2)
    
    if den == 0:
        return 0
    
    else:
        corr = num/den
        return corr

def get_patch_centres(img, win_size=25, stride=128):
    '''
    Function to get patch centres
    
    Input
    -----
    img: Input image
    win_size: Window size of the patch
    stride: Stride with which the window moves in both x and y axes
    
    Output
    ------
    centres: Array of the patch centres
    
    '''
    mid = win_size // 2
    offset = win_size % 2
    h, w, c = img.shape
    centres = []
    
    for i in range(mid, w - mid, stride):
        for j in range(mid, h - mid, stride):
            centres.append((j, i))
            
    return np.asarray(centres)

def int_win_corr(img1, img2, win_size=25, stride=20):
    '''
    Function to find matches using intensity window based correlation
    
    Input
    -----
    img1: Input image 1
    img2: Input image 2
    win_size: Window size
    stride: Stride with which window moves
    
    Output
    ------
    matches1: Matches in the image1
    matches2: Matches in the image2
    
    '''
    pts1 = get_patch_centres(img1, win_size, stride)
    pts2 = get_patch_centres(img2, win_size, stride)
    
    matches1 = []
    matches2 = []
    
    w = win_size // 2
    off = win_size % 2
    
    for i in pts1:
        max_cor = 0
        max_patch_centre = None
        patch1 = img1[i[0]-w:i[0]+w+off, i[1]-w:i[1]+w+off]
        
        for j in pts2:
            patch2 = img2[j[0]-w:j[0]+w+off, j[1]-w:j[1]+w+off]
            corr = correlation(patch1, patch2)
            if corr > max_cor:
                max_cor = corr
                max_patch_centre = j
                
        if max_patch_centre is not None:
            matches1.append(i)
            matches2.append(max_patch_centre)
        
    return np.int32(matches1), np.int32(matches2)
```

## 2.2 Recall that for a point in one image, its corresponding point in the other image is found along its epipolar line. Select few points in the first image and then plot their corresponding epipolar lines on the second. Repeat the same for the second image.

```python
def drawlines(img1,img2,lines,pts1,pts2):
    '''
    Function to draw lines
    
    Input
    -----
    img1: Input image 1
    img2: Input image 2
    lines: Lines
    pts1: Selected match coordinates in image 1
    pts2: Selected match coordinates in image 2
    
    Output
    ------
    img1: Output image 1 with epipolar lines
    img2: Output image 2 with epipolar lines
    
    '''
    r,c,_ = img1.shape
    for r,pt1,pt2 in zip(lines,pts1,pts2):
        color = tuple(np.random.randint(0,255,3).tolist())
        x0,y0 = map(int, [0, -r[2]/r[1] ])
        x1,y1 = map(int, [c, -(r[2]+r[0]*c)/r[1] ])
        img1 = cv2.line(img1, (x0,y0), (x1,y1), color,1)
        img1 = cv2.circle(img1,tuple(pt1),5,color,-1)
        img2 = cv2.circle(img2,tuple(pt2),5,color,-1)
    return img1,img2
```

```python
imgs = import_imgs(1)

kp1, des1 = get_keypoints(imgs[0])
kp2, des2 = get_keypoints(imgs[1])

matches = matching(kp1, des1, kp2, des2)
pts1, pts2 = get_match_coord(matches, kp1, kp2)
pts1 = np.int32(pts1)
pts2 = np.int32(pts2)

F, mask = cv2.findFundamentalMat(pts1, pts2, cv2.FM_LMEDS)

pts1 = pts1[mask.ravel() == 1]
pts2 = pts2[mask.ravel() == 1]
```

```python
def get_epipolarLines(coords, F):
    coords = np.hstack((coords, np.ones(coords.shape[0]).reshape(-1, 1)))
    M = np.matmul(coords, F)
    return M/(M[:,2].reshape(-1,1))
```

## 2.3 Rectify the pair of images and on these new images, find the correspondences using greedy matching

```python
def stereo_rectify(img1, img2, win_size=64, stride=64):
    '''
    Function to do stereo rectification of the stereo images
    
    Input
    -----
    img1: Input stereo image 1
    img2: Input stereo image 2
    win_size: Window size
    stride: Stride with which window moves
    
    Output
    ------
    new_img1: Stereo rectified image 1
    new_img2: Stereo rectified image 2
    
    '''
    match1, match2 = int_win_corr(img1, img2, win_size, stride)
    
    F, mask = cv2.findFundamentalMat(match1, match2, cv2.FM_LMEDS)
    
    _, H1, H2 = cv2.stereoRectifyUncalibrated(match1, match2, F, img1.shape[:2])
    
    h, w, c = img1.shape
    new_img1 = cv2.warpPerspective(img1, H1, (w, h))
    new_img2 = cv2.warpPerspective(img2, H2, (w, h))
    
    return new_img1, new_img2
```

```python
imgs = import_imgs(1)

new1, new2 = stereo_rectify(imgs[0], imgs[1])

plt.imshow(new1)
plt.show()

plt.imshow(new2)
plt.show()
```

## 2.4 Compare the Brute force and Greedy based solution for the stereo correspondence problem.

* Brute force method searches exhaustively over all patches
* Greedy based method search only on the epipolar line
* Greedy based method works much faster because search space is reduced
* Greedy based method is based on the concept that the corresponding point of a point lies on epipolar line
