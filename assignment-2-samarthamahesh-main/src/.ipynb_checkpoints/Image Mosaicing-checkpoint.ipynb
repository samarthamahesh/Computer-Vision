```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Helper Functions

```python
def import_imgs(idx):
    '''
    Function to import a set of images
    
    Input
    -----
    idx: Index of the image
    
    Output
    ------
    imgs: Array of images
    
    '''
    imgs = []
    idx = str(idx)
    ilen = len(idx)
    path = '../images/Image Mosaicing/'
    
    for i in os.listdir(path):
        if i[:ilen] == idx:
            img = cv2.imread(path+i)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            imgs.append(img)
            
    return imgs
```

## 1.1 Use any feature detector and descriptor (e.g. SIFT) to find matches between two partially overlapping images. You can use inbuilt functions for this.

```python
def get_keypoints(im):
    '''
    Function to find keypoints and descriptors using feature detector and descriptor (SIFT)
    
    Input
    -----
    im: Image
    
    Output
    ------
    kp: Keypoints
    des: Descriptor
    
    '''
    sift = cv2.SIFT_create()
    kp, des = sift.detectAndCompute(im, None)
    
    return kp, des

def matching(kp1, des1, kp2, des2, ratio=0.5):
    '''
    Function tp find matches between two images using SIFT outputs
    
    Input
    -----
    kp1: Keypoints of image 1
    des1: Descriptor of image 1
    kp2: Keypoints of image 2
    des2: Descriptor of image 2
    ratio: Paramater to decide the match based on distances
    
    Output
    ------
    new_matches: Matches obtained between two images
    
    '''
    FLANN_INDEX_KDTREE = 1
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks=50) 

    flann = cv2.FlannBasedMatcher(index_params, search_params)

    matches = flann.knnMatch(des1, des2, k=2)

    new_matches = []
    for i, (m, n) in enumerate(matches):
        if m.distance < ratio * n.distance:
            new_matches.append([m.queryIdx, m.trainIdx])
    
    return np.asarray(new_matches)

def get_match_coord(matches, kp1, kp2):
    '''
    Function to get coordinates of the matches obtained
    
    Input
    -----
    matches: Matches indices
    
    Output
    ------
    coord1: Coordinates of the match points in image 1
    coord2: Coordinates of the match points in image 2
    
    '''
    coord1 = []
    coord2 = []
    for i, j in matches:
        coord1.append(kp1[i].pt)
        coord2.append(kp2[j].pt)
        
    return np.asarray(coord1), np.asarray(coord2)

def show_matches(img1, img2, kp1, kp2, matches):
    drawn_matches = cv2.drawMatches(img1, kp1, img2, kp2, matches, img2, flags=2)
    plt.figure(figsize=(25,25))
    plt.imshow(drawn_matches)
    plt.show()
```

## 1.2 Estimate the homography matrix between the two images robustly.

```python
def DLT_homography(im1, im2):
    '''
    Function to obtain homography matrix using DLT method
    
    Input
    -----
    im1: Image 1 coordinates
    im2: Image 2 coordinates
    
    Output
    ------
    H: Homography matrix
    
    '''
    G = []
    for i in range(im1.shape[0]):
        x = im2[i][0]
        y = im2[i][1]
    
        u = im1[i][0]
        v = im1[i][1]
        
        G.append([x, y, 1, 0, 0, 0, -u*x, -u*y, -u])
        G.append([0, 0, 0, x, y, 1, -v*x, -v*y, -v])

    U, S, Vh = np.linalg.svd(G)
    
    H = Vh[-1, :] / Vh[-1, -1]
    H = H.reshape(3, 3)
    
    return H

def project_err(H, im1, im2):
    '''
    Function to find error
    
    Input
    -----
    H: Homography matrix
    im1: Image 1 coordinates
    im2: Image 2 coordinates
    
    Output
    ------
    error: Projection error
    
    '''
    im1_proj = H @ np.append(im2, 1)
    im1_proj /= im1_proj[2]
        
    error = np.mean(np.square(im1[0] - im1_proj[:2]))
    
    return error

def RANSAC_homography(im1, im2, iters, thresh):
    '''
    Function to find homography matrix using RANSAC method
    
    Input
    -----
    im1: Image 1 coordinates
    im2: Image 2 coordinates
    iters: Number of iterations
    thresh: Threshold for projection error
    
    Output
    ------
    H_out: Best homography matrix
    best_err: Best projection error
    max_inliers: Maximum number of inliers
    
    '''
    max_inliers = 0
    best_err = 0
    H_out = np.zeros((3, 4))
    
    for i in range(iters):
        ind = np.random.randint(0, im2.shape[0], 4)
        sample_im1 = im1[ind]
        sample_im2 = im2[ind]
        H = DLT_homography(sample_im1, sample_im2)
        
        err_sum = 0
        inliers = 0
        for j in range(im2.shape[0]):
            im1_p = im1[j:j+1]
            im2_p = im2[j:j+1]
            
            err = project_err(H, im1_p, im2_p)
            err_sum += err
            if err < thresh:
                inliers += 1
        
        if inliers > max_inliers:
            H_out = H
            best_err = err_sum / im2.shape[0]
            max_inliers = inliers
            
    return H_out, best_err, max_inliers
```

## 1.3 Transform one of the images to the otherâ€™s reference frame using the homography matrix.

```python
def project(H, img1, img2):
    '''
    Function to project image to other image's reference frame
    
    Input
    -----
    H: Homography matrix
    img: Image
    
    Output
    ------
    ret: Projected image
    
    '''
    ret = cv2.warpPerspective(img2, H, (img1.shape[1]+img2.shape[1], (img1.shape[0]+img2.shape[0])))
    
    return ret
```

```python
imgs = import_imgs(2)

kp1, des1 = get_keypoints(imgs[0])
kp2, des2 = get_keypoints(imgs[1])

matches = matching(kp1, des1, kp2, des2)
coord1, coord2 = get_match_coord(matches, kp1, kp2)

H, _, _ = RANSAC_homography(coord1, coord2, 100, 10)

ret = project(H, imgs[0], imgs[1])

plt.imshow(ret)
plt.show()
```

## 1.4 Stitch the two images together.

```python
def stitch(img1, img2, H):
    '''
    Function to stitch two images
    
    Input
    -----
    img1: Image 1
    img2: Image 2
    H: Homography matrix
    
    Output
    ------
    result: Output stitched image
    
    '''
    h1, w1, _ = img1.shape
    h2, w2, _ = img2.shape
    
    bp1 = np.array([[[0, 0]], [[0, h1]], [[w1, h1]], [[w1, 0]]], dtype='float32')
    bp2 = np.array([[[0, 0]], [[0, h2]], [[w2, h2]], [[w2, 0]]], dtype='float32')
    
    tmp = cv2.perspectiveTransform(bp2, H)
    bp = np.concatenate((bp1, tmp), axis=0)
    
    xmin, ymin = np.int32(np.min(bp, axis=0).reshape(-1))
    xmax, ymax = np.int32(np.max(bp, axis=0).reshape(-1))

    diff = (xmax-xmin, ymax-ymin)
    t = [-xmin, -ymin]
    Ht = np.array([[1, 0, t[0]], [0, 1, t[1]], [0, 0, 1]])

    result = cv2.warpPerspective(img2, Ht@H, diff)
    
    mask = np.where(img1 > 0, True, False)
    result[t[1]:h1+t[1], t[0]:w1+t[0]] = mask*img1 + (1-mask)*result[t[1]:h1+t[1], t[0]:w1+t[0]]
    
    return result
```

## 1.5 Repeat this for multiple images to produce a singly mosaic/panorama.

```python
def stitch_multiple(imgs):
    '''
    Function to stitch all images in a set
    
    Input
    -----
    imgs: Array of images
    
    Output
    ------
    res: Output stitched panaromic image
    
    '''
    res = imgs[0]
    
    for i in imgs:
        kp1, des1 = get_keypoints(res)
        kp2, des2 = get_keypoints(i)

        matches = matching(kp1, des1, kp2, des2)
        coord1, coord2 = get_match_coord(matches, kp1, kp2)
        H, err, inl = RANSAC_homography(coord1, coord2, 100, 0.01)

        res = stitch(res, i, H)
    
    return res
```
